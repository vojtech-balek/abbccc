```python
import yfinance as yf
import pandas as pd

# Define the copper futures ticker (HG=F for Copper Futures)
copper_ticker = "HG=F"

# Download historical data for copper using yfinance
# Adjust the period and interval as needed (e.g., 5y for 5 years of data)
copper_data = yf.download(copper_ticker, period="5y", interval="1d")

# Copper futures are traded in cents per pound, so convert to USD per kilogram:
# 1 pound = 0.453592 kg and price is in cents, so multiply by (100 / 0.453592)
copper_data['Price_per_kg'] = copper_data['Adj Close'] / 0.453592

# Display the first few rows of the modified data
print(copper_data[['Price_per_kg']].head())

# Save the data to a CSV file (optional)
copper_data[['Price_per_kg']].to_csv("copper_price_per_kg.csv", index=True)
copper_data.to_csv("copper_price.csv")
# Plot the price per kg (optional, requires matplotlib)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(copper_data.index, copper_data['Price_per_kg'], label='Copper Price per kg (USD)', color='blue')
plt.title('Copper Price per Kilogram (Historical Data)')
plt.xlabel('Date')
plt.ylabel('Price (USD per kg)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the aluminium futures ticker (ALI=F for Aluminium Futures)
aluminium_ticker = "ALI=F"

# Download historical data for aluminium using yfinance
# Adjust the period and interval as needed (e.g., 5y for 5 years of data)
aluminium_data = yf.download(aluminium_ticker, period="5y", interval="1d")

# Aluminium futures are typically priced in USD per metric ton.
# Convert price from metric tons to kilograms (1 metric ton = 1000 kg).
# Therefore, to get price per kg, we divide by 1000.
aluminium_data['Price_per_kg'] = aluminium_data['Adj Close'] / 1000

# Display the first few rows of the modified data
print(aluminium_data[['Price_per_kg']].head())

# Save the data to a CSV file (optional)
aluminium_data[['Price_per_kg']].to_csv("aluminium_price_per_kg.csv", index=True)
aluminium_data.to_csv("aluminium_price.csv")
# Plot the price per kg (optional)
plt.figure(figsize=(10, 6))
plt.plot(aluminium_data.index, aluminium_data['Price_per_kg'], label='Aluminium Price per kg (USD)', color='orange')
plt.title('Aluminium Price per Kilogram (Historical Data)')
plt.xlabel('Date')
plt.ylabel('Price (USD per kg)')
plt.grid(True)
plt.legend()
plt.show()
```
